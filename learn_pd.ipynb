```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
%cd C:\Users\Katie\Documents\GitHub\learn_python
bowl = pd.read_csv("data/kaitlin-dio-history.csv")
```

```python
bowl.drop(labels = bowl.tail(1).index, inplace = True)

display(bowl.tail(10))
```

```python
bowl.describe()
```

# Data Visualization

## Matplotlib

```python
plt.scatter(bowl['Week'], bowl['Todays Avg'])
plt.title('Scatter Plot of Average by Week')

plt.xlabel('Week')
plt.xlabel('Average')
```

```python
plt.plot(bowl['Week'], bowl['Todays Avg'])

plt.title('Plot of Average by Week')

plt.xlabel('Week')
plt.ylabel('Average')
```

## Seaborn

```python
bowl_wd = pd.wide_to_long(df = bowl, stubnames = 'Gm', i = 'Week', j = 'Game')
bowl_wd.head(10)
```

```python
cols_plot = ["Gm1", "Gm2", "Gm3", "Todays Avg"]

ax = sns.lineplot(data = bowl[cols_plot])

ax.set(xlabel='Week', ylabel='Score')
```

```python
ax = sns.lineplot(data = bowl_wd, x = 'Week', y = 'Gm')

ax.set(xlabel='Week', ylabel='Score')
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
```
